```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_trueNews = pd.read_csv("../data/True.csv")
df_fakeNews = pd.read_csv("../data/Fake.csv")
```

```python
df_trueNews.head()
```

```python
df_fakeNews.head()
```

```python
real_count = len(df_trueNews)
fake_count = len(df_fakeNews)

labels = ['Real News', 'Fake News']
counts = [real_count, fake_count]

plt.figure()
plt.bar(labels, counts, color=['blue', 'red'])

plt.title("Fake vs Real News Distribution")
plt.ylabel("Number of Articles")

plt.tight_layout()
```

```python
true_counts = df_trueNews['subject'].value_counts()
true_counts
```

```python
fake_counts = df_fakeNews['subject'].value_counts()
fake_counts
```

```python
plt.figure()
true_counts.plot(kind='bar', color='blue')
plt.title("True News Subject Distribution")
plt.xlabel("Subject")
plt.ylabel("Count")
```

```python
plt.figure()
fake_counts.plot(kind='bar', color='red')
plt.title("Fake News Subject Distribution")
plt.xlabel("Subject")
plt.ylabel("Count")
```

```python
df_trueNews['length'] = df_trueNews['text'].apply(len)

plt.figure()
plt.hist(df_trueNews['length'], bins=50, label='Real', color='blue')

plt.legend()
plt.title("True News Text Length Distribution")
plt.xlabel("Length")
plt.ylabel("Frequency")
```

```python

df_fakeNews['length'] = df_fakeNews['text'].apply(len)
plt.figure()
plt.hist(df_fakeNews['length'], bins=50, label='Fake', color="red")
plt.legend()
plt.title("Fake News Text Length Distribution")
plt.xlabel("Length")
plt.ylabel("Frequency")
```
